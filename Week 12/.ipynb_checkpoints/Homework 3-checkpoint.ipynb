```python
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier

x = [0, 1, 2, 3]

y = [0, 0, 0, 0]

classes = [0, 0, 1, 1]

data = list(zip(x, y))

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(data, classes)

new_x = 1.1
new_y = 0

new_point_1 = [(new_x, new_y)]

new_x_2 = 1.6
new_y_2 = 0

new_point_2 = [(new_x_2, new_y_2)]

prediction = knn.predict(new_point_1)

plt.scatter(x + [new_x], y + [new_y], c=classes + [prediction[0]])
plt.text(x=new_x-0.5, y=new_y-0.01, s=f"new point, class: {prediction[0]}")
plt.show()

prediction = knn.predict(new_point_2)

plt.scatter(x + [new_x_2], y + [new_y_2], c=classes + [prediction[0]])
plt.text(x=new_x_2-0.5, y=new_y_2-0.01, s=f"new point, class: {prediction[0]}")
plt.show()
```
