Create an array containing 250 random floats between 0 and 5:

```python
import numpy
import matplotlib as mat
import matplotlib.pyplot as plt

x = numpy.random.uniform(0, 5, 250)

#print(x)
```

```python
plt.hist(x, 5)

plt.show()
```

We use the array from the example above to draw a histogram with 5 bars.

The first bar represents how many values in the array are between 0 and 1.

The second bar represents how many values are between 1 and 2.

NORMAL DATA DISTRIBUTION:
In the previous examples we learned how to create a completely random array, of a given size, and between two given values.

In this chapter we will learn how to create an array where the values are concentrated around a given value.

In probability theory, this kind of data distribution is known as the normal data distribution, or the Gaussian data distribution

```python
random_numbers_normally_distributed = numpy.random.normal(5, 1, 100000) # The mean value is 5, and the standard deviation is 1. 
                                                                        #Give me an array with 100000 values in it.


plt.hist(random_numbers_normally_distributed, 100) # to draw a histogram with 100 bars

plt.show()
```

We use the array from the numpy.random.normal() method, with 100000 values, to draw a histogram with 100 bars.

We specify that the mean value is 5, and the standard deviation is 1.

Meaning that the values should be concentrated around 5, and rarely further away than 1 from the mean.

And as you can see from the histogram, most values are between 4 and 6, with a top at approximately 5.

SCATTER PLOT: A scatter plot is a diagram where each value in the data set is represented by a dot.

The Matplotlib module has a method for drawing scatter plots, it needs two arrays of the same length, 
one for the values of the x-axis, and one for the values of the y-axis.
    

```python
age_of_cars = [5,7,8,7,2,17,2,9,4,11,12,9,6]

speed_of_cars = [99,86,87,88,111,86,103,87,94,78,77,85,86]
```

The age_of_cars array represents the age of each car.

The speed_of_cars array represents the speed of each car.

Use the scatter() method to draw a scatter plot diagram.

```python
plt.scatter(age_of_cars, speed_of_cars)
```

Scatter Plot Explained:

The x-axis represents ages, and the y-axis represents speeds.

What we can read from the diagram is that the two fastest cars were both 2 years old, and the slowest car was 12 years old.

Note: It seems that the newer the car, the faster it drives, but that could be a coincidence, after all we only registered 13 cars.

Random Data Distributions:

Let us create two arrays that are both filled with 1000 random numbers from a normal data distribution.

The first array will have the mean set to 5 with a standard deviation of 1.

The second array will have the mean set to 10 with a standard deviation of 2.

So, A scatter plot with 1000 dots will be created:

```python
x_values = numpy.random.normal(5, 1, 1000) # Give me 1000 random numbers around 5 with standard deviation of 1.

y_values = numpy.random.normal(10, 2, 1000) # Give me 1000 random numbers around 10 with standard deviation of 2.

plt.scatter (x_values, y_values) # A scatter plot with 1000 dots.
plt.show()
```

Scatter Plot Explained:

We can see that the dots are concentrated around the value 5 on the x-axis, and 10 on the y-axis.

We can also see that the spread is wider on the y-axis than on the x-axis.
